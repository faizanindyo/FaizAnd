```python
import pandas as pd

# Membaca file CSV dan membuat DataFrame
df = pd.read_csv('/content/drive/MyDrive/MACHINE LAERNING/diamonds.csv')

# Menampilkan beberapa baris pertama dari

print(df.head())
```

```python

print(df.describe())
```

```python
import seaborn as sns

# Memilih kolom-kolom numerik untuk diplot
numeric_cols = df.select_dtypes(include=['float64', 'int64'])

# Membuat boxplot menggunakan seaborn
sns.boxplot(data=numeric_cols)
```

```python

import matplotlib.pyplot as plt


# Memilih kolom-kolom numerik untuk diplot
numeric_cols = df.select_dtypes(include=['float64', 'int64'])

# Menampilkan boxplot untuk setiap fitur numerik
for column in numeric_cols.columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=column, data=df)
    plt.title(f'Boxplot of {column}')
    plt.show()
```

```python
# Memilih kolom-kolom numerik untuk diplot
numeric_cols = df.select_dtypes(include=['float64', 'int64'])

# Menampilkan boxplot untuk setiap fitur numerik
for column in numeric_cols.columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=column, data=df)
    plt.title(f'Boxplot of {column}')
    plt.show()


```

```python
# Memilih kolom-kolom numerik untuk diplot
numeric_cols = df.select_dtypes(include=['float64', 'int64'])

# Membuat scatter plot untuk setiap pasangan fitur numerik
sns.pairplot(df[numeric_cols.columns])
plt.show()
```

```python
# Memilih kolom-kolom numerik untuk diplot
numeric_cols = df.select_dtypes(include=['float64', 'int64'])

# Menghitung korelasi antara fitur-fitur numerik
correlation_matrix = numeric_cols.corr()

# Membuat heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Heatmap of Numeric Features Correlation')
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=['float64', 'int64'])

# Menampilkan distribusi dari setiap fitur numerik
for column in numeric_cols.columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
numeric_cols = df.select_dtypes(include=['float64', 'int64'])

# Menampilkan violin plot dari setiap fitur numerik
for column in numeric_cols.columns:
    plt.figure(figsize=(8, 6))
    sns.violinplot(y=df[column])
    plt.title(f'Violin Plot of {column}')
    plt.ylabel(column)
    plt.show()

```
