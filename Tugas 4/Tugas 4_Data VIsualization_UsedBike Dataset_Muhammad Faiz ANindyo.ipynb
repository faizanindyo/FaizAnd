```python
import pandas as pd

# Membaca file CSV
df = pd.read_csv("/content/drive/MyDrive/MACHINE LAERNING/Used_Bikes.csv")

# Menampilkan lima baris pertama dari DataFrame
print(df.head())
```

```python
# Menampilkan statistik fitur
print(df.describe())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Memilih kolom-kolom numerik
numerical_columns = ['price', 'kms_driven', 'age', 'power']

# Menampilkan boxplot untuk setiap fitur numerik menggunakan seaborn dan matplotlib
for column in numerical_columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=df, y=column)
    plt.title(f"Boxplot of {column}")
    plt.ylabel("Values")
    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Memilih kolom-kolom numerik
numerical_columns = ['price', 'kms_driven', 'age', 'power']

# Membuat pairplot untuk semua fitur numerik
sns.pairplot(df[numerical_columns])
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt



# Memilih kolom-kolom numerik
numerical_columns =  ['price', 'kms_driven', 'age', 'power']

# Menghitung korelasi antar fitur numerik
correlation_matrix = df[numerical_columns].corr()

# Membuat heatmap dari korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Heatmap of Numerical Features')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Memilih kolom-kolom numerik
numerical_columns =  ['price', 'kms_driven', 'age', 'power']

# Menampilkan distribusi dari setiap fitur numerik
for column in numerical_columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.grid(False)
    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Memilih kolom-kolom numerik
numerical_columns =  ['price', 'kms_driven', 'age', 'power']


# Menampilkan violin plot dari setiap fitur numerik
for column in numerical_columns:
    plt.figure(figsize=(8, 6))
    sns.violinplot(data=df, y=column)
    plt.title(f'Violin Plot of {column}')
    plt.ylabel('Values')
    plt.grid(False)
    plt.show()
```
