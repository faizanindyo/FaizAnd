```python
# Nama: [Masukkan nama lengkap Anda]
# NIM: [Masukkan nomor induk mahasiswa Anda]
# Judul Tugas: Machine Learning dengan Linear Regression pada Dataset Salary_Data_Simple_SVR.csv
# Lecture: [Masukkan nomor atau judul lecture tempat tugas ini diberikan]
# Model: Linear Regression
# Dataset: Salary_Data_Simple_SVR.csv

# Import library yang diperlukan
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load dataset dari file CSV
data = pd.read_csv('/content/drive/MyDrive/Tugas 2 Machine Learning/Salary_Data_Simple_SVR.csv')

# Melihat lima baris pertama dari dataset untuk memahami struktur data
print(data.head())

# Memisahkan fitur (X) dan target (y)
X = data.iloc[:, :-1].values
y = data.iloc[:, 1].values

# Membagi dataset menjadi data latih dan data uji dengan perbandingan 80:20
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Membuat objek model regresi linear
model = LinearRegression()

# Melatih model menggunakan data latih
model.fit(X_train, y_train)

# Memprediksi nilai gaji menggunakan data uji
y_pred = model.predict(X_test)

# Membuat plot untuk visualisasi data latih dan model
plt.scatter(X_train, y_train, color='blue')
plt.plot(X_train, model.predict(X_train), color='red')
plt.title('Gaji vs Pengalaman (Data Latih)')
plt.xlabel('Pengalaman (Tahun)')
plt.ylabel('Gaji ($)')
plt.show()

# Membuat plot untuk visualisasi data uji dan prediksi model
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_train, model.predict(X_train), color='red')
plt.title('Gaji vs Pengalaman (Data Uji)')
plt.xlabel('Pengalaman (Tahun)')
plt.ylabel('Gaji ($)')
plt.show()

# Menghitung nilai Mean Squared Error (MSE) untuk evaluasi model
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

# Menampilkan koefisien regresi (slope)
print("Koefisien Regresi (Slope):", model.coef_)

# Menampilkan intercept (nilai konstanta)
print("Intercept (Konstanta):", model.intercept_)
```
